```python
import pandas as pd
import numpy as np
import random
```

```python
class MyPCA:
    def __init__(self, n_components=3):
        self.n_components = n_components

    def __str__(self):
        return f"MyPCA class: n_components={self.n_components}"

    def fit_transform(self, X: pd.DataFrame):
        means = X.mean(axis=0)
        X = (X - means) 
        cov_matrix = X.T @ X / (len(X) -1 )
        
        eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
        
        sorted_values_and_vectors = sorted(zip(eigenvalues, eigenvectors.T), key=lambda x: abs(x[0]), reverse=True)
        vectors = np.array([vector for value, vector in sorted_values_and_vectors[:self.n_components] ])
        return X @ (vectors.T)
```

```python
X = pd.DataFrame({'a': [1,2,3,4], 'b': [9,8,7,6], 'c': [2,7,4,9]})
```

```python
MyPCA().fit_transform(X)
```
