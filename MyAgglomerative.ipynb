```python
import pandas as pd
import numpy as np
import random
```

```python


class MyAgglomerative:
    
    def __init__(self, n_clusters=3, metric="euclidean"):
        self.n_clusters = n_clusters
        self.metric = metric
        
    def __str__(self):
        return f"MyAgglomerative class: n_clusters={self.n_clusters}"

    def fit_predict(self, X: pd.DataFrame):
        clusters_indexes = [ [index] for index in X.index]

        while len(clusters_indexes) > self.n_clusters:
        
            min_distance = np.inf
            min_i = None
            min_j = None
            
            for i in range(1, len(clusters_indexes)):
                for j in range(i):
                    X_i = X.loc[clusters_indexes[i]].mean(axis=0)
                    X_j = X.loc[clusters_indexes[j]].mean(axis=0)
    
                    difference = X_i - X_j

                    if self.metric == "euclidean":
                        distance = (difference ** 2).sum() ** 0.5

                    if self.metric == 'chebyshev':
                        distance = difference.abs().max()
                    if self.metric == 'manhattan':
                        distance = difference.abs().sum()
                    if self.metric == 'cosine':
                        distance = 1 - (X_i * X_j).sum() / ( np.sqrt((X_i ** 2).sum()) * ((X_j ** 2).sum() ** 0.5) )
    
                    if distance < min_distance:
                        min_distance = distance
                        min_i = i
                        min_j = j

            clusters_indexes.append(clusters_indexes[min_i] + clusters_indexes[min_j])
            del clusters_indexes[min_i]
            del clusters_indexes[min_j]

        y = []
        
        for cluster_index, cluster_indexes in enumerate(clusters_indexes):
            y.append(pd.Series(cluster_index, cluster_indexes))

        y = pd.concat(y)
        return y
                
```

```python
X = pd.DataFrame({'a': [1,2,3,5,6,7,9,10]})
```

```python
MyAgglomerative().fit_predict(X)
```
